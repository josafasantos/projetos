# Recomendação de Filmes

```python
# importando as bibliotecas
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# Carrega o dataset
df = pd.read_csv('../../datasets/ratings.csv')
```

```python
# Verifica informações basicas
df.head()
```

```python
df.info()
```

```python
# Cria uma tabela de Pivot
matriz_usu_item = df.pivot_table(
  index='user_id',
  columns='item_id',
  values='rating'
)
```

```python
matriz_usu_item
```

```python
matriz_usu_item = matriz_usu_item.fillna(0)
```

```python
# Calcular similaridade entre Usuários
similaridade_usuarios = cosine_similarity(matriz_usu_item)
```

```python
# Transforma em DataFrame para facilitar manipulação
similaridade_usuarios_df = pd.DataFrame(
  similaridade_usuarios, index=matriz_usu_item.index,
  columns=matriz_usu_item.index
)
```

```python
# Cria função de recomendação
def recomendar_filmes(usuario_alvo, num_recom=5):
  # Ver a similidade do usuario alvo com todos os outros
  notas_usuario_alvo = matriz_usu_item.loc[usuario_alvo]
  similaridades = similaridade_usuarios_df.loc[usuario_alvo]

  # Para cada item calcular uma nota prevista

  notas_previstas = {}

  for item in matriz_usu_item.columns:
    if notas_usuario_alvo[item] == 0:
      soma_similaridade = 0
      soma_poderada = 0
      for outro_usuario in matriz_usu_item.index:
        if matriz_usu_item.loc[outro_usuario, item] > 0:
          sim = similaridades[outro_usuario]
          rating_outro_user = matriz_usu_item.loc[outro_usuario, item]
          soma_poderada += sim * rating_outro_user
          soma_similaridade += sim
      
      if soma_similaridade > 0:
        nota_final = soma_poderada / soma_similaridade
      else:
        nota_final = 0
      notas_previstas[item] = nota_final

  # Ordenar pelos itens com maior nota prevista
  recomenda = sorted(notas_previstas.items(), key=lambda x:x[1], reverse=True)

  return recomenda[:num_recom]
```

```python
# Exemplo de recomendação
usuario_exemplo = 3
recomendacoes = recomendar_filmes(usuario_exemplo, num_recom=3)
print(f"Recomendações para o usuário {usuario_exemplo}")
for item , nota in recomendacoes:
  print(f"Item: {item} | Nota Prevista: {nota:.2f}")
```
