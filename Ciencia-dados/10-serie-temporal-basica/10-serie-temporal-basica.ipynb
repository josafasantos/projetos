# Serie Temporal

```python
# Importando as bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Carregar e preparar os dados

```python
# Carregar o dataset de temperatura diaria
df_temperatura = pd.read_csv("../../datasets/temperatura_diaria_exemplo.csv")
```

```python
# Converte a coluna "Data" para tipo DateTime
df_temperatura['Data'] = pd.to_datetime(df_temperatura['Data'])
```

### Visualização

```python
# Cria o gráfico de Série Temporal
plt.figure(figsize=(12,6))
plt.plot(df_temperatura['Data'], df_temperatura['Temperatura_C'], label="Temperatura Diária (°C)")

# Adiciona Linha de Tendência - Média Movel de 30 dias
periodo_media_movel = 30
df_temperatura['Media_Movel_30D'] = df_temperatura['Temperatura_C'].rolling(window=periodo_media_movel, center=True).mean()
plt.plot(df_temperatura['Data'], df_temperatura['Media_Movel_30D'], color='red', linestyle='--', label=f'Média Móvel ({periodo_media_movel} dias)')

#Personalização do grafico
plt.title("Evolução da Temperatura Diária ao Longo do Tempo")
plt.xlabel('Data')
plt.ylabel('Temperatura (°C)')
plt.legend()
plt.grid(True, linestyle=":")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
